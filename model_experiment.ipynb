```python
df = pd.read_csv(PATH + 'train.csv')
df.shape
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=1603)
train.shape, test.shape
```

```python
x_train = train.drop(['Id', 'SalePrice'], axis=1)
x_test = test.drop(['Id', 'SalePrice'], axis=1)
y_train = train['SalePrice']
y_test = test['SalePrice']
```

```python
cat_cols = x_train.select_dtypes(include='object').columns
num_cols = x_train.select_dtypes(exclude='object').columns

print(f"Categorical columns ({len(cat_cols)}): {cat_cols}")
print(f"Numerical columns ({len(num_cols)}): {num_cols}")
```

## პირველი მარტივი მოდელი

### feature engineering

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn import set_config

set_config(transform_output="pandas")

# ჩავანაცვლოთ რიცხვები საშუალოთი, კატეგორიები რიცხვებით (NaN ან უცნობი -> -1)
preprocessor = ColumnTransformer([
    ("num", SimpleImputer(strategy="mean"), num_cols),
    ("cat", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1, encoded_missing_value=-1), cat_cols)
])
preprocessor.set_output(transform="pandas")

x_train = pd.DataFrame(
    preprocessor.fit_transform(train),
    columns=preprocessor.get_feature_names_out()
)

x_test = pd.DataFrame(
    preprocessor.transform(test),
    columns=preprocessor.get_feature_names_out()
)
```

```python
x_train.head()

```

```python
import mlflow
import mlflow.sklearn
import dagshub

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

REPO_OWNER = 'adola23'
REPO_NAME = 'HousePricesML'

dagshub.init(repo_owner=REPO_OWNER, repo_name=REPO_NAME, mlflow=True)
```

### პირველი მარტივი მოდელის ექსპერიმენტი

```python
# პირველი მარტივი მოდელის ლოგი
mlflow.set_experiment("BasicLinearRegression1")
with mlflow.start_run():

    model = LinearRegression()
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    y_pred = np.maximum(0, y_pred)
    y_pred_log = np.log1p(y_pred)
    y_test_log = np.log1p(y_test)
    rmse_log = np.sqrt(mean_squared_error(y_test_log, y_pred_log))

    mlflow.log_param("model_type", "LinearRegression")
    mlflow.log_param("test_size", 0.2)


    mlflow.log_metric("rmse_log", rmse_log)
    mlflow.log_metric("mse", mse)
    mlflow.log_metric("r2", r2)


    mlflow.sklearn.log_model(model, "model")
    mlflow.sklearn.log_model(preprocessor, "preprocessor")

    print("RMSE:", rmse_log)
    print("MSE:", mse)
    print("R2:", r2)
```

## უკეთესი მოდელის feature engineering და ექსპერიმენტი ერთად

```python
# ახლა ვცადოთ რიცხვები ჩავანაცვლოთ საშუალოებით, კატეგორიულები თუ #კატეგორია <= one_hot_threshold მაშინ one hot encoding 
# თუ > 3 მაშინ woe და ბოლოს კორელაციის ფილტრს მოვდებთ (ეს აღარ არის კოდში, iv + woe მოვაშორე რადგან კარგი შედეგი არ მომცა)

# აქ არის one_hot + კორელაციის ფილტრი + ლოგარითმზე ტრენინგი რაც rmse-ს მინიმიზაციაში გვეხმარება

x_train = train.drop(['Id', 'SalePrice'], axis=1)
x_test = test.drop(['Id', 'SalePrice'], axis=1)
y_train = train['SalePrice']
y_test = test['SalePrice']

```

```python
import joblib # ამას ვიყენებ რომ იგივე პრეპროცესინგი გავაკეთო model_inference-ში
```

```python
X_train, X_test = x_train.copy(), x_test.copy()
# ეს kaggle-ს საიტზე ეწერა რომ შესაძლოა კარგი ყოფილიყო
train_idx = X_train[X_train['GrLivArea'] < 4000].index
X_train, y_train = X_train.loc[train_idx], y_train.loc[train_idx]


mlflow.set_experiment("LinearRegression ONE_HOT +  CORRELATION + train on log(y)")
with mlflow.start_run():
    corr_threshold = 0.99
    one_hot_threshold = 30

    mlflow.log_param("correlation_threshold", corr_threshold)
    mlflow.log_param("one_hot_threshold", one_hot_threshold)

    # ======== Feature engineering ===========

    train_means = {col: x_train[col].mean() for col in num_cols}

    # რიცხვები საშუალოთი ჩავანაცვლოთ
    for col in num_cols:
        mean_value = X_train[col].mean()
        X_train[col].fillna(mean_value, inplace=True)
        X_test[col].fillna(mean_value, inplace=True)

    # კატეგორიულები
    one_hot_cols_list = []
    one_hot_columns = {}
    one_hot_used_cols = []
    chosen_for_one_hot = 0
    for col in cat_cols:
        n_unique = X_train[col].nunique()
        if n_unique <= one_hot_threshold:
            # one hot encoding
            dummies_train = pd.get_dummies(X_train[col], prefix=col)
            dummies_test = pd.get_dummies(X_test[col], prefix=col)

            dummies_test = dummies_test.reindex(columns=dummies_train.columns, fill_value=0)

            one_hot_columns[col] = list(dummies_train.columns)
            one_hot_used_cols.append(col)
            
            X_train = pd.concat([X_train.drop(columns=[col]), dummies_train], axis=1)
            X_test = pd.concat([X_test.drop(columns=[col]), dummies_test], axis=1)

            
            one_hot_cols_list += list(dummies_train.columns)
            chosen_for_one_hot += 1
    

    # ======== Feature Selection ===========

    mlflow.log_param("dropped_cat_columns", len(cat_cols) - chosen_for_one_hot)
    # დავტოვოთ შესაბამისი სვეტები
    selected_cols = list(num_cols) + one_hot_cols_list
    X_train_filtered = X_train[selected_cols].copy()
    X_test_filtered = X_test[selected_cols].copy()

    # კორელაციის ფილტრი
    corr_matrix = X_train_filtered.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    to_drop = [col for col in upper.columns if any(upper[col] > corr_threshold)]
    mlflow.log_dict({"dropped_columns": to_drop}, "correlation_dropped.json")

    X_train_filtered.drop(columns=to_drop, inplace=True)
    X_test_filtered.drop(columns=to_drop, inplace=True)


    # პრეპროცესინგის მონაცემების შენახვა
    preprocessing_artifacts = {
        "num_cols": num_cols,
        "cat_cols": cat_cols,
        "train_means": train_means,
        "one_hot_used_cols": one_hot_used_cols,
        "one_hot_columns": one_hot_columns,
        "selected_cols": selected_cols,
        "to_drop": to_drop
    }
    joblib.dump(preprocessing_artifacts, "preprocessing.pkl")
    mlflow.log_artifact("preprocessing.pkl")

    # ======== Model Training ===========
    
    # რადგან საბოლოო მეტრიკა kaggle-ზე არის rmse 
    # მოდელს დავატრენინგებ y-ის ლოგარითმზე 
    y_train_log = np.log1p(y_train)
    y_test_log = np.log1p(y_test)

    model = LinearRegression()

    model.fit(X_train_filtered, y_train_log)

    y_pred_log = model.predict(X_test_filtered)
    y_train_pred_log = model.predict(X_train_filtered)

    rmse_log_train = np.sqrt(mean_squared_error(y_train_log, y_train_pred_log))
    rmse_log = np.sqrt(mean_squared_error(y_test_log, y_pred_log))

    mlflow.log_param("model_type", "LinearRegression")
    mlflow.log_param("test_size", 0.2)


    mlflow.log_metric("rmse_log", rmse_log)


    mlflow.sklearn.log_model(model, "model")

    print("RMSE:", rmse_log)
    print("TRAIN RMSE:", rmse_log_train)
```

## მესამე ექსპერიმენტი

## feature engineering

```python
# scaler-ის გამოყენების დამატება და დამატებითი column-ების დამატება 
# სხვადასხვა წყვილებით რომ უფრო მეტი კავშირი დაინახოს მოდელმა

X_train, X_test = x_train.copy(), x_test.copy()
train_idx = X_train[X_train['GrLivArea'] < 4000].index
X_train, y_train = X_train.loc[train_idx], y_train.loc[train_idx]

Y_train = y_train.copy()
Y_test = y_test.copy()

Y_train_log = np.log1p(y_train)
Y_test_log = np.log1p(y_test)

# დავამატოთ თითოეული სვეტისთვის ცალკე სვეტი რომელიც გვეტყვის ეს სვეტი არის თუ არა NaN
for col in num_cols:
    if X_train[col].isna().sum() > 0:
        X_train[f"{col}_isnan"] = X_train[col].isna().astype(int)
        X_test[f"{col}_isnan"] = X_test[col].isna().astype(int)



# რიცხვები საშუალოთი ჩავანაცვლოთ
for col in num_cols:
    mean_value = X_train[col].mean()
    X_train[col].fillna(mean_value, inplace=True)
    X_test[col].fillna(mean_value, inplace=True)


# დავამატოთ სვეტების ლოგარითმებიც რადგან შესაძლოა ლოგარითმის პოვნაში უკეთ გამოგვადგეს 
for col in num_cols:
    if (X_train[col] > 0).all():
        X_train[f"{col}_log"] = np.log1p(X_train[col])
        X_test[f"{col}_log"] = np.log1p(X_test[col])


col_x_col_count = 10
# ავიღოთ ათი პასუხთან ყველაზე კორელირებული სვეტი და მათი წყვილები დავამატოთ
corr = X_train[num_cols].corrwith(y_train).abs().sort_values(ascending=False)
top_cols = corr.head(col_x_col_count).index

for i in range(len(top_cols)):
    for j in range(i+1, len(top_cols)):
        c1 = top_cols[i]
        c2 = top_cols[j]
        X_train[f"{c1}_x_{c2}"] = X_train[c1] * X_train[c2]
        X_test[f"{c1}_x_{c2}"] = X_test[c1] * X_test[c2]
```

```python
ohe_dummies = {}
# one hot encoding
for col in cat_cols:
    # one hot encoding
    dummies_train = pd.get_dummies(X_train[col], prefix=col)
    dummies_test = pd.get_dummies(X_test[col], prefix=col)
    ohe_dummies[col] = dummies_train.columns
    dummies_test = dummies_test.reindex(columns=dummies_train.columns, fill_value=0)
    X_train = pd.concat([X_train.drop(columns=[col]), dummies_train], axis=1)
    X_test = pd.concat([X_test.drop(columns=[col]), dummies_test], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# scaler-ს აღარ ვიყენებ რადგან უარეს შედეგს მაძლევდა. 
X_train_scaled = X_train
X_test_scaled  = X_test
```

### Feature Selection

```python
# RFE

from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

lr = LinearRegression()

rfe_feature_count = 100
n_features_to_select = min(rfe_feature_count, X_train_scaled.shape[1])

rfe = RFE(estimator=lr, n_features_to_select=n_features_to_select)

X_rfe = rfe.fit_transform(X_train_scaled, Y_train_log)

# keep selected feature names
selected_features = X_train_scaled.columns[rfe.support_]
print("Selected features:", list(selected_features))
```

```python
X_train_filtered = X_train_scaled[selected_features]
X_test_filtered = X_test_scaled[selected_features]


# პრეპროცესინგის მონაცემების შენახვა
preprocessing_artifacts = {
    "num_cols": num_cols,
    "cat_cols": cat_cols,
    "train_means": train_means,
    "top_cols": top_cols,
    "ohe_dummies": ohe_dummies,
    "selected_features": selected_features,
}
joblib.dump(preprocessing_artifacts, "preprocessing.pkl")

```

### train + ექსპერიმენტი

```python
mlflow.set_experiment("LinearRegression scaling + rfe + logs + col_X_col + one_hot")
with mlflow.start_run():

    mlflow.log_param("col_X_col top col count", col_x_col_count)
    mlflow.log_param("rfe feature count", rfe_feature_count)
    mlflow.log_param("selected_features", selected_features)

    model = LinearRegression()

    model.fit(X_train_filtered, Y_train_log)

    Y_pred_log = model.predict(X_test_filtered)

    rmse_log = np.sqrt(mean_squared_error(Y_test_log, Y_pred_log))

    mlflow.log_param("model_type", "LinearRegression")
    mlflow.log_param("test_size", 0.2)


    mlflow.log_metric("rmse_log", rmse_log)


    mlflow.sklearn.log_model(model, "model")
    mlflow.log_artifact("preprocessing.pkl")
    
    print("RMSE:", rmse_log)
```

# მეოთხე ექსპერიმენტი წინას preprocessing-ით

```python
from sklearn.tree import DecisionTreeRegressor

mlflow.set_experiment("DecisionTree rfe + logs + col_X_col + one_hot")
with mlflow.start_run():

    mlflow.log_param("col_X_col top col count", col_x_col_count)
    mlflow.log_param("rfe feature count", rfe_feature_count)
    mlflow.log_param("selected_features", selected_features)

    model = DecisionTreeRegressor(
        random_state=42,
        max_depth=5,
        min_samples_split=30,
        min_samples_leaf=20,
    )   

    model.fit(X_train_filtered, Y_train_log)

    Y_pred_log = model.predict(X_test_filtered)
    Y_train_pred_log = model.predict(X_train_filtered)

    rmse_log = np.sqrt(mean_squared_error(Y_test_log, Y_pred_log))
    rmse_log_train = np.sqrt(mean_squared_error(Y_train_log, Y_train_pred_log))
    mlflow.log_param("model_type", "DecisionTreeRegressor")
    mlflow.log_param("test_size", 0.2)

    mlflow.log_param("max_depth", model.max_depth)
    mlflow.log_param("min_samples_split", model.min_samples_split)
    mlflow.log_param("min_samples_leaf", model.min_samples_leaf)
    mlflow.log_param("random_state", 42)

    mlflow.log_metric("rmse_log", rmse_log)

    mlflow.sklearn.log_model(model, "model")
    mlflow.log_artifact("preprocessing.pkl")

    print("RMSE:", rmse_log)
    print("TRAIN RMSE:", rmse_log_train)
```
